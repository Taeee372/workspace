```python
# 리스트에 데이터 추가 
# 1. append() 함수 사용 : 리스트의 마지막 요소에 데이터가 추가
list1 = []
list1.append(1)
print(list1[0])
list1.append(2)
list1.append(3)
print(list1)

# 2. insert(삽입할 index, 데이터) : 원하는 위치에 데이터를 추가
list1.insert(1, 4)
print(list1)
```

```python
# 리스트 데이터 삭제
list1 = [1, 2, 3, 4, 5]

del list1[2] # list1 변수의 2번 요소 데이터를 삭제
print(list1)

# 주의 사항!
del list1 # list1 변수 자체를 제거! (빈 배열이 되는 게 아님)
# print(list1) list1 변수 자체가 지워지기 때문에 출력 시 오류!
```

```python
# 리스트 데이터 수정
list1 = [1, 2, 3]

# 2를 5로 변경
list1[1] = 5
print(list1)
```

```python
# 슬라이싱 : 리스트에서 특정 영역만 추출하는 문법
# [추출시작 index : 추출 마지막 index(마지막 인덱스 전까지의 값이 나온다)]
list2 = [1, 2, 3, 4, 5]
print(list2[1:4]) # 1번 index에서 4번 index 전까지!
print(list2[:3]) # 시작 인텍스를 생략하면 0번째 인덱스로 간주
print(list2[2:])
print(list2[:])
print(list2[2:-1])
```
