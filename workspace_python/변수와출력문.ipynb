### 출력문
파이썬에서 출력문은 print() 명령어를 사용합니다.

```python
# 이것은 주석입니다.
```

```python
# 파이썬 명령어는 마지막에 세미콜론 붙이지 않음!
print(10)
print('문자는 홀따옴표에 감싼다')
print("문자는 쌍따옴표도 가능하다")
```

```python
# 연산 결과의 출력
print(10 + 5)
print(10 / 4) # 연산 결과가 실수면 실수값이 나옴
print(10 // 4) # 나눗셈의 몫
print(10 % 4) # 나눗셈의 나머지
print('java' + 'python')
# print('python' + 5) 파이썬은 숫자와 문자의 더하기 안 됨(python5가 아니라 걍 오류남)
```

```python
# print() 활용
print(10, 'hello', 10.5)

# print() 함수는 'end' 매개변수로 출력 후 실행 코드를 작성할 수 있음
print('hi', end='\n')
print('hi', end='')
print('hi', end='.')
```

### 변수
python은 javascript와 마찬가지로 내부적으로만 자료형이 존재한다.

```python
# 변수의 선언 및 초기화
num = 10
print(num)
```

```python
myName = '홍길동'
myAge = 20

# 변수와 함께 쓰려면 '' 앞에 f 붙이기
# 이름은 홍길동이고, 나이는 20살입니다
print(f'이름은 {myName}이고, 나이는 {myAge}살 입니다')
```
